post.likes: postì˜ likeê°œìˆ˜ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤ 

post.get_comments() : ëª¨ë“  commentë¥¼ ì–»ìŠµë‹ˆë‹¤ iteration í˜•íƒœë¡œ ì–»ê¸° ë•Œë¬¸ì— ì´ ì•ˆì—ì„œ íŠ¹ì • ì •ë³´ë¥¼ ë½‘ì•„ë‚´ì„œ ì¶”ê°€ì‹œì¼œì•¼í•  ê²ƒ ê°™ìŠµë‹ˆë‹¤

post.get_likes() : ì¢‹ì•„ìš”ë¥¼ ëˆ„ë¥¸ userë“¤ì„ ë³´ì—¬ì¤ë‹ˆë‹¤ í•˜ì§€ë§Œ ì´ ì—­ì‹œ iteratorì´ê¸° ë•Œë¬¸ì— ì¶”ê°€ì‹œí‚¤ë©´ì„œ ë„£ì–´ì•¼í•  ê²ƒ ê°™ìŠµë‹ˆë‹¤ 

getcomments()ë¥¼í†µí•´ì„œ  answersë¥¼ ëŒë¦¬ë©´ì„œ ê°€ì§€ê³  ìˆëŠ” ëª¨ë“  answerë“¤ì„ ë½‘ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. profileì„ ì œì™¸í•˜ê³  ìš°ì„  ë°ì´í„°ë¥¼ ë½‘ì•„ë³´ë„ë¡ í•©ì‹œë‹¤! 


```python
import pandas as pd 
import tqdm

i=1
content_list=[]
for post in tqdm.tqdm(profile.get_posts()):
    post_list.append(post)
    com_list=[]
    for com in post.get_comments():
        ans_list=[]
        for ans in com.answers:
            ans_list.append([ans.text,ans.owner,ans.likes_count])
        com_list.append([com.text,com.likes_count,com.owner,ans_list])
    content_list.append([i,post.date,post.profile,post.typename,post.caption_hashtags,post.is_video,post.video_view_count,post.video_duration,post.likes,post.comments,com_list])
    
    i+=1
    
df=pd.DataFrame(content_list,columns=["id","date","profile","type","caption_and_hashtags","is_video","view_count","video_duration","likes","comments","com_list"])
df.head()
df.to_csv("hilton_post2",encoding="utf-8")
```
