```python
def extract_text(text):
    '''comment Ï∂îÏ∂ú Ìï®Ïàò'''
    regex=re.compile(r"\'.*\'")
    mo=regex.search(text)
    if mo!=None:
        return mo.group()
    else:
        return ""

def extract_profile(text):
    '''profile id Ï∂îÏ∂ú Ìï®Ïàò'''
    regex=re.compile("\(.*\)")
    mo=regex.search(text)
    if mo!=None:
        return mo.group()[1:-1]
    else:
        return ""
```

```python
# comment Ï∂îÏ∂ú 
hilton["comment"]=hilton["com_list"].apply(lambda x:extract_text(x.split(",")[0])[1:-1])
```

```python
# \\nÏ†úÍ±∞ Î¨∏ÏûêÎßå ÎÇ®Í≤®Ï§çÎãàÎã§
hilton['comment']=hilton['comment'].apply(lambda x: x.strip(".\\n"))
hilton['comment']
```

```python
# postÍ∞Ä Î∞õÏùÄ Ï¢ãÏïÑÏöî Í∞úÏàò 
hilton['comment_like']=hilton['com_list'].apply(lambda x: int(x.split(",")[1]))
hilton['comment_like'].head(10)
```

```python
hilton['comment_like'].unique()
```

```python
# comment ÏûëÏÑ±Ïûê 
hilton["comment_profile"]=hilton["com_list"].apply(lambda x: extract_profile(x.split(",")[2]))
hilton['comment_profile']
```

```python
hilton['comment_profile'].unique()
```

```python
# ans_listÎ•º Îã§Ïãú ÎΩëÏïÑÎ¥ÖÎãàÎã§
hilton['ans_list']=hilton['com_list'].apply(lambda x:x.split(",")[3:])# Ï∂îÌõÑ Ï†ïÎ¶¨Ìï† ÏòàÏ†ï 
hilton.drop(['com_list'],axis=1,inplace=True)
hilton.head()
```
